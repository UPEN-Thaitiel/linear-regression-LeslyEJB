#### Movies per year

```python
movies_per_year = df.groupby('year')['name'].count()
movies_per_year
```

```python
plt.plot(movies_per_year.index, movies_per_year.values, marker='o', color='y')
plt.title("Number of movies per year")
plt.xlabel("Year")
plt.ylabel("Number of movies")
plt.grid(True)
plt.show()
```

#### Avg budget vs Gross over the years

```python
budget_gross = df.groupby('year')[['budget','gross']].mean()
budget_gross
```

```python
plt.plot(budget_gross.index, budget_gross['budget'], marker='o', color='red', label='budget')
plt.plot(budget_gross.index, budget_gross['gross'], marker='x', color='g', label='gross')
plt.title("Budget vs Gross")
plt.grid(True)
plt.legend()
```

#### Correlation

#### Import libraries

```python
# Linear Regression
from sklearn.linear_model import LinearRegression

# Sample split
#from sklearn.model_selection import train_test_split

# Model Evaluation
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

##### Separate predictor from target variable (independent vs dependent)

```python
X = df[["budget"]]
y = df[["gross"]]
```

##### Train the model

```python
model = LinearRegression()
model.fit(X, y)
```

##### Use the model to predict values

```python
y_pred = model.predict(X)
```

##### Model Evaluation

```python
r2 = r2_score(y, y_pred)
print("R2: ", r2)
```

```python
plt.scatter(X, y, alpha=0.5)
plt.plot(X, y_pred, color="Red")
plt.show()
```
